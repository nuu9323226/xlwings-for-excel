## 2018年及2019年報表表格整合

### 一、設定方式

1.設定檔案名稱、2018分頁名稱、2019分頁名稱、合併的分頁名稱、要整理的月份報表名稱<br >

2.欄位名稱必須為｛18/01數量｝ 月份與數量中間不可以有空格｛18/01 數量｝ <br >
  要有資訊包含數量及金額、值不能為空白，才可以完整執行<br >
  範例如下:<br >
sheet 2018<br >

| 18碼 | 18/01數量 | 18/02數量 | 18數量合計 | 18/01金額 | 18/02金額 | 18金額合計 |
| --- | --- | --- |
| VCF108M010S0ANH03L | 14000 | 50000 | 64000 | 672 | 2375 | 3047 |

sheet 2019<br >

| 18碼 | 19/01數量 | 19/02數量 | 19數量合計 | 19/01金額 | 19/02金額 | 19金額合計 |
| --- | --- | --- |
| VCF108M010S0ANH03L | 16000 | 40000 | 64000 | 675 | 2372 | 3047 |

3.如果設定3月到5月，則sheet 2018、sheet 2019皆要有3月到5月的資料，值為0亦可，但不能沒有該月份欄位及值

4.第一欄必須為18碼<====最重要

```python
import xlwings as xw
import re
from fnmatch import fnmatch, fnmatchcase

def Type(items):
    type=[]
    for a in items:
        if a[0:1]=='J':
            type.append(Typecode[a[0:1]])
        elif a[0:1]=='T':
            type.append(Typecode[a[0:1]])
        elif a[0:1]=='N':
            type.append(Typecode[a[0:1]])
        elif a[0:1]=='V':
            type.append(Typecode[a[0:1]])
        else:
            Unknow=a[0:1]
            unk='unkonw Tpye: %s' %(Unknow)
            print('unkonw tpye: %s' %(Unknow))
            type.append(unk)
    return type

# def Voltage(items):

#     voltage=[]
#     for a in items:
# #         print(a)
#         new=a.split('M',1)
#         VAL=new[1][0:3]
# #         print('valtage: %s'%(VAL))
#         if fnmatch(VAL, '*R*'):
#             newnum=VAL.split('R',1)
#             pointnum=newnum[0]+'.'+newnum[1]
# #             print('valtage: %s'%(pointnum))
#             voltage.append(pointnum)
#         else:
#             voltage.append(str(int(VAL)))
#     return voltage
# #             print('valtage: %s'%(int(VAL)))


def Voltage(items):

    voltage=[]
    
    for a in items:
#         print(a)
        VAL=a[7:10]
#         print('valtage: %s'%(VAL))
        if fnmatch(VAL, '*R*'):
            newnum=VAL.split('R',1)
            pointnum=newnum[0]+'.'+newnum[1]
    #           print('valtage: %s'%(pointnum))
            voltage.append(pointnum)
        else:
            try: 
                voltage.append(str(int(VAL))) 
            except:
                print('worning:%s its not value of voltage' %(a))
                voltage.append('none') 
    return voltage



def Size(items):
    casesize=[]
    caselength=[]
    for a in items:
        newsize=a[-3:-1]
#         print(newsize)
        try:
            if a[0:1]=='J':
#                 print('JTNsize: ',JTNlengthcode[newsize])
                caselength.append(JTNlengthcode[newsize])
            elif a[0:1]=='T':
#                 print('JTNsize: ',JTNlengthcode[newsize])
                caselength.append(JTNlengthcode[newsize])
            elif a[0:1]=='N':
#                 print('JTNsize: ',JTNlengthcode[newsize])
                caselength.append(JTNlengthcode[newsize])
            elif a[0:1]=='V':
#                 print('Vsize: ',Vlengthcode[newsize])
                caselength.append(Vlengthcode[newsize])
            else:
                unk='unkonw Length:%s' %(newsize)
                print(unk)
                caselength.append(unk)
        except:
            print('unkonw Length:%s' %(newsize))
            unk='unkonw Length:%s' %(newsize)
            caselength.append(unk)
        
        try:
            newDiameter=a[-4:-3]
#             print('Diametercode: ',Diametercode[newDiameter])
            casesize.append(Diametercode[newDiameter])
        except:
            print('unkonw Ø:%s' %(newDiameter))
            unk='unkonw Ø:%s' %(newDiameter)
            casesize.append(unk)
    return casesize,caselength

def capacitorsize(items):
    capsize=[]
    for a in items:
        newcap=a[3:5]
#         print('vold:',newcap)
        unit=a[5:6]
#         print('unit',unit)
        try:
            finalcap=float(newcap)*float(capacitor[unit])
#         print(str(finalcap))
#         task='%fuF' %(finalcap)
            capsize.append(finalcap)
        except:
            print('worning:%s its not value of voltage' %(a))
            capsize.append('none')
    return capsize
    
    
    
def partnumber_main(pathfile,sheetfile):
    wb = xw.Book(pathfile)
    sheet1 = wb.sheets[sheetfile]

    rng=sheet1.range('A1').expand('table')

    nrows=rng.rows.count

    ncols=rng.columns.count
    print(nrows)
    print(ncols)
    b=sheet1.range('A2:A%d' %(nrows)).value
    print(b)
    type=Type(b)
    print(type)
    voltage=Voltage(b)
    print(voltage)

    casesize,caselength=Size(b)
    print(casesize)
    print(caselength)
    capvalue=capacitorsize(b)


    # print(str(len(b)))
    sheet1.range('B:F').api.Insert()
    sheet1.cells(1,'B').value='Type'
    sheet1.cells(1,'C').value='uF'
    sheet1.cells(1,'D').value='Voltage'
    sheet1.cells(1,'E').value='Ø'
    sheet1.cells(1,'F').value='Length'
    for i in range(len(b)):
        sheet1.cells(i+2,'B').value=type[i]
        sheet1.cells(i+2,'C').value=str(capvalue[i])
        sheet1.cells(i+2,'D').value=voltage[i]
        sheet1.cells(i+2,'E').value=casesize[i]
        sheet1.cells(i+2,'F').value=caselength[i]

    print('============finish=============')
    
def deleteDuplicatedElementFromList3(listA):
    return sorted(set(listA), key = listA.index)
# https://www.cnblogs.com/tudas/p/python-delete-duplicate-element-from-list.html

def partnumberCombine(partnumber18,partnumber19):
    s118=deleteDuplicatedElementFromList3(partnumber18)
    s119=deleteDuplicatedElementFromList3(partnumber19)
#     print('2018 after:  ',s118)
#     print('2019 after:   ',s119)
    partnumber1819=s118+s119
    s1819=deleteDuplicatedElementFromList3(partnumber1819)
#     print(s1819)
    return s1819
```

```python
wb = xw.Book(pathfile)
sheet2018 = wb.sheets[sheet2018file]
sheet2019 = wb.sheets[sheet2019file]

# 新增分頁，若讀取不到則增加頁面
try:
    sheetnew = wb.sheets[sheetnewfile]

except:
    wb.sheets.add(sheetnewfile)
    print('add sheet: ',sheetnewfile)
sheetnew = wb.sheets[sheetnewfile]
# sheetnew.cells(1, 1).value = "Hello World!"
sheetnew.clear()
#獲得列長度
rng18=sheet2018.range('A1').expand('table')
nrows18=rng18.rows.count
partnumber18=sheet2018.range('A2:A%d' %(nrows18)).value
print('2018 row:',nrows18)
# print('2018:  ',partnumber18)

rng19=sheet2019.range('A1').expand('table')
nrows19=rng19.rows.count
partnumber19=sheet2019.range('A2:A%d' %(nrows19)).value
print('2019 row:',nrows19)
# print('2019:   ',partnumber19)


#獲得唯一沒有重複的partnumber
Com1819=partnumberCombine(partnumber18,partnumber19)
print('partnumberCombine',Com1819)


#獲得2018的欄標題
ncols18=rng18.columns.count
# colsline18=sheet2018.range('A1:A%d' %(nrows18)).value
# sheetnew.range('a1:d4').value = [1,2,3,4,5,6]
fst_col18 = sheet2018[0,:ncols18].value
print('fst_col18',fst_col18)


#獲得2019的欄標題
ncols19=rng19.columns.count
# colsline18=sheet2018.range('A1:A%d' %(nrows18)).value
# sheetnew.range('a1:d4').value = [1,2,3,4,5,6]
fst_col19 = sheet2019[0,:ncols19].value
print('fst_col19',fst_col19)




# 在新的sheet建構新的欄位標題
print(str(len(Com1819)))
newfullq=full.copy()
sheetnew.cells(1,'A').value='18碼'
# newfullq.remove('10')
# newfullq.remove('11')
# newfullq.remove('12')
print('newfullq',newfullq)

# 將18/ 數量文字加上去
newfull=[]
for a in newfullq:
    abc='18/'+a+'數量'
    newfull.append(abc)
newfull.append('18數量合計')
for a in newfullq:
    abc='18/'+a+'金額'
    newfull.append(abc)
newfull.append('18金額合計')

for a in full:
    abc='19/'+a+'數量'
    newfull.append(abc)
newfull.append('19數量合計')
for a in full:
    abc='19/'+a+'金額'
    newfull.append(abc)
newfull.append('19金額合計')

print(newfull)



# 18/ 數量文字加上文字後再貼在新表格的第一欄
sheetnew.range('b1:d%d' %(len(newfull))).value = newfull



# 建構出來後讀取第一排為陣列fst_colnew
rngnew=sheetnew.range('A1').expand('table')
ncolsnew=rngnew.columns.count
fst_colnew = sheetnew[0,:ncolsnew].value
print('fst_colnew',fst_colnew)
    


for partid in range(len(Com1819)):
    #將18碼貼在新的表格的第一欄
    sheetnew.cells(partid+2,'A').value=Com1819[partid]
    totalnum=0
    totalmoney=0
    if Com1819[partid] in partnumber18 : #如果料號有在sheet2018中則進行複製貼上
        for i in newfullq:
            #橫的
            indextitle=fst_col18.index('18/%s數量' %(i))
#             print('在2018該月份[]的位置[]',i,indextitle+1)
            #直向
            indextitleid=partnumber18.index(Com1819[partid])
#             print('在2018 partnumber[]的位置[]',Com1819[partid],indextitleid+1)

#             sheetnew.cells(partid+2,'B').value =sheet2018.cells(indextitleid+2,indextitle+1).value
            title_new=fst_colnew.index('18/%s數量' %(i))
            sheetnew.cells(partid+2, title_new+1  ).value =sheet2018.cells(indextitleid+2,indextitle+1).value
            totalnum=totalnum+sheet2018.cells(indextitleid+2,indextitle+1).value
#         print('totalnum',totalnum)    
        title_new1=fst_colnew.index('18數量合計')    
        sheetnew.cells(partid+2, title_new1+1  ).value =totalnum
        
        
        for i in newfullq:
            #橫的
            indextitle=fst_col18.index('18/%s金額' %(i))
#             print('在2018該月份[]的位置[]',i,indextitle+1)
            #直向
            indextitleid=partnumber18.index(Com1819[partid])
#             print('在2018 partnumber[]的位置[]',Com1819[partid],indextitleid+1)

#             sheetnew.cells(partid+2,'B').value =sheet2018.cells(indextitleid+2,indextitle+1).value
            title_new=fst_colnew.index('18/%s金額' %(i))
            sheetnew.cells(partid+2, title_new+1  ).value =sheet2018.cells(indextitleid+2,indextitle+1).value
            totalmoney=totalmoney+sheet2018.cells(indextitleid+2,indextitle+1).value
#         print('totalmoney',totalmoney)    
        title_new1=fst_colnew.index('18金額合計')    
        sheetnew.cells(partid+2, title_new1+1  ).value =totalmoney
        
        
        
#         2019==================================================================
    totalnum19=0
    totalmoney19=0
    if Com1819[partid] in partnumber19 : #如果料號有在sheet2018中則進行複製貼上
        for i in full:
            #橫的
            indextitle=fst_col19.index('19/%s數量' %(i))
#             print('在2019該月份[]的位置[]',i,indextitle+1)
            #直向
            indextitleid=partnumber19.index(Com1819[partid])
#             print('在2019 partnumber[]的位置[]',Com1819[partid],indextitleid+1)

#             sheetnew.cells(partid+2,'B').value =sheet2018.cells(indextitleid+2,indextitle+1).value
            title_new=fst_colnew.index('19/%s數量' %(i))
            sheetnew.cells(partid+2, title_new+1  ).value =sheet2019.cells(indextitleid+2,indextitle+1).value
            totalnum19=totalnum19+sheet2019.cells(indextitleid+2,indextitle+1).value
#         print('totalnum19',totalnum19)    
        title_new1=fst_colnew.index('19數量合計')    
        sheetnew.cells(partid+2, title_new1+1  ).value =totalnum19
        
        
        for i in full:
            #橫的
            indextitle=fst_col19.index('19/%s金額' %(i))
#             print('在2019該月份[]的位置[]',i,indextitle+1)
            #直向
            indextitleid=partnumber19.index(Com1819[partid])
#             print('在2019 partnumber[]的位置[]',Com1819[partid],indextitleid+1)

#             sheetnew.cells(partid+2,'B').value =sheet2018.cells(indextitleid+2,indextitle+1).value
            title_new=fst_colnew.index('19/%s金額' %(i))
            sheetnew.cells(partid+2, title_new+1  ).value =sheet2019.cells(indextitleid+2,indextitle+1).value
            totalmoney19=totalmoney19+sheet2019.cells(indextitleid+2,indextitle+1).value
#         print('totalmoney19',totalmoney19)    
        title_new1=fst_colnew.index('19金額合計')    
        sheetnew.cells(partid+2, title_new1+1  ).value =totalmoney19

print('============finish=============')
```
