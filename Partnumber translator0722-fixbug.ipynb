### 二、若有下面例外請確認

1.目前內建T:Snap In, V:SMD , J:Radial , N: Screw這幾個型號的電容，如果需要新增，請在此處修改Typecode = {   }，新增格式如範例 'V': 'SMD'<br >

2.如果有J,T,N,V以外的開頭的電容Length，會因為不再下表資料內，會顯示unknow Length，需修改code<br >

```python
def Type(items):
    type=[]
    for a in items:
        if a[0:1]=='J':
            type.append(Typecode[a[0:1]])
        elif a[0:1]=='T':
            type.append(Typecode[a[0:1]])
        elif a[0:1]=='N':
            type.append(Typecode[a[0:1]])
        elif a[0:1]=='V':
            type.append(Typecode[a[0:1]])
        else:
            Unknow=a[0:1]
            unk='unkonw Tpye: %s' %(Unknow)
            print('unkonw tpye: %s' %(Unknow))
            type.append(unk)
    return type

# def Voltage(items):

#     voltage=[]
#     for a in items:
# #         print(a)
#         new=a.split('M',1)
#         VAL=new[1][0:3]
# #         print('valtage: %s'%(VAL))
#         if fnmatch(VAL, '*R*'):
#             newnum=VAL.split('R',1)
#             pointnum=newnum[0]+'.'+newnum[1]
# #             print('valtage: %s'%(pointnum))
#             voltage.append(pointnum)
#         else:
#             voltage.append(str(int(VAL)))
#     return voltage
# #             print('valtage: %s'%(int(VAL)))


def Voltage(items):

    voltage=[]
    
    for a in items:
#         print(a)
        VAL=a[7:10]
#         print('valtage: %s'%(VAL))
        if fnmatch(VAL, '*R*'):
            newnum=VAL.split('R',1)
            pointnum=newnum[0]+'.'+newnum[1]
    #           print('valtage: %s'%(pointnum))
            voltage.append(pointnum)
        else:
            try: 
                voltage.append(str(int(VAL))) 
            except:
                print('worning:%s its not value of voltage' %(a))
                voltage.append('none') 
    return voltage



def Size(items):
    casesize=[]
    caselength=[]
    for a in items:
        newsize=a[-3:-1]
#         print(newsize)
        try:
            if a[0:1]=='J':
#                 print('JTNsize: ',JTNlengthcode[newsize])
                caselength.append(JTNlengthcode[newsize])
            elif a[0:1]=='T':
#                 print('JTNsize: ',JTNlengthcode[newsize])
                caselength.append(JTNlengthcode[newsize])
            elif a[0:1]=='N':
#                 print('JTNsize: ',JTNlengthcode[newsize])
                caselength.append(JTNlengthcode[newsize])
            elif a[0:1]=='V':
#                 print('Vsize: ',Vlengthcode[newsize])
                caselength.append(Vlengthcode[newsize])
            else:
                unk='unkonw Length:%s' %(newsize)
                print(unk)
                caselength.append(unk)
        except:
            print('unkonw Length:%s' %(newsize))
            unk='unkonw Length:%s' %(newsize)
            caselength.append(unk)
        
        try:
            newDiameter=a[-4:-3]
#             print('Diametercode: ',Diametercode[newDiameter])
            casesize.append(Diametercode[newDiameter])
        except:
            print('unkonw Ø:%s' %(newDiameter))
            unk='unkonw Ø:%s' %(newDiameter)
            casesize.append(unk)
    return casesize,caselength

def capacitorsize(items):
    capsize=[]
    for a in items:
        newcap=a[3:5]
#         print('vold:',newcap)
        unit=a[5:6]
#         print('unit',unit)
        try:
            finalcap=float(newcap)*float(capacitor[unit])
#         print(str(finalcap))
#         task='%fuF' %(finalcap)
            capsize.append(finalcap)
        except:
            print('worning:%s its not value of voltage' %(a))
            capsize.append('none')
    return capsize
```

```python
wb = xw.Book(pathfile)
sheet1 = wb.sheets[sheetfile]

rng=sheet1.range('A1').expand('table')

nrows=rng.rows.count

ncols=rng.columns.count
print(nrows)
print(ncols)
b=sheet1.range('A2:A%d' %(nrows)).value
print(b)
type=Type(b)
print(type)
voltage=Voltage(b)
print(voltage)

casesize,caselength=Size(b)
print(casesize)
print(caselength)
capvalue=capacitorsize(b)


# print(str(len(b)))
sheet1.range('B:F').api.Insert()
sheet1.cells(1,'B').value='Type'
sheet1.cells(1,'C').value='uF'
sheet1.cells(1,'D').value='Voltage'
sheet1.cells(1,'E').value='Ø'
sheet1.cells(1,'F').value='Length'
for i in range(len(b)):
    sheet1.cells(i+2,'B').value=type[i]
    sheet1.cells(i+2,'C').value=str(capvalue[i])
    sheet1.cells(i+2,'D').value=voltage[i]
    sheet1.cells(i+2,'E').value=casesize[i]
    sheet1.cells(i+2,'F').value=caselength[i]
    
print('============finish=============')
```
